```python
### This cell imports the data from a text file that contains the distances between features and sorts them
### by chromosome. It creates a histogram for each chromosome that shows the distances between the features. 

%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import os
import scipy
import sys
from scipy import stats
import matplotlib.ticker as mtick


### Initiate lists of variables for later use
all_distances = []
all_data = dict()
def main():  
    ### Pick the file directory
    directory = '/media/ab/data/Research_Data/'
    
    ### Pick the input files
    distance_file = 'div_class_notx_bidir_k1.bed' 
    chr_size_file = 'mm10.chrom.sizes.txt'
    
    dir_list = sorted(os.listdir(directory))
    
    chr_sizes = load_chr_sizes(directory+chr_size_file)
    print chr_sizes
    for filename in dir_list:
        if filename == distance_file:      ### Load only the TSV files
            all_data[filename] = load_from_file(directory+filename, chr_sizes)
            print "length of all data: ", len(all_data)

    for key in sorted(all_data.keys()):      
        print "key: ", key
        make_graphs(key, all_data[key])

def load_chr_sizes(filename):
    chr_sizes = dict()
    x=open(filename)   ### Open the file
    for line in x:               ### For each line in the file 
        print line
        chromosome = (line.strip('\n').split('\t')[0])
        print "chromosome: ", chromosome
        size = int(line.strip('\n').split('\t')[1])/1000000
        #print "distance: ", distance
        chr_sizes[chromosome] = size
    x.close()
    return chr_sizes

def load_from_file(file_to_open, chr_sizes):              
        
    distances = dict()
    x=open(file_to_open)   ### Open the file
    data = []                    ### Initiate the data list
    for line in x:               ### For each line in the file 
        chromosome = (line.strip('\n').split('\t')[3])
        #print "chromosome: ", chromosome
        distance = (line.strip('\n').split('\t')[6])
        #print "distance: ", distance
        if chromosome in distances.keys():
            distances[chromosome].append((int(distance)/chr_sizes[chromosome]))
        else:
            if chromosome != '.':
                distances[chromosome] = [(int(distance)/chr_sizes[chromosome])]
                print "chromosome: ", chromosome
    x.close()
    return distances 
        
        
def make_graphs(filename, data):
    
    #print "x1 length : ", len(x1)
    
 
    #print x1_bins, x2_bins, x3_bins, x4_bins, x5_bins

    ### Make the five histogram axes for the plot
    fig, axs = plt.subplots(nrows=10, ncols=2, figsize=(12, 12))
    
    axs = axs.ravel()
    
    #print chrom
    j = 0
    for key in (data.keys()):
        ### Calculate the number of bins for the histogram
        print "len(data[key]) :", len(data[key]), "key: ", key
        #x_bins = len(data[key]) / 20
        x_bins = 500
        ### Add the data to the histograms
        axs[j].hist(data[key], x_bins, normed=1, histtype='bar')
        axs[j].set_title(key)
        axs[j].locator_params(axis='x',nbins=3)
        axs[j].set_yscale("log", nonposy='clip')
        axs[j].ticklabel_format(style='sci', axis='x', scilimits=(0,0))
        axs[j].set_xlim(left=-5000, right=5000)
        j += 1
        print "j: ", j
    
    
    ### Add a title to the plot
    fig.suptitle(filename, y=1.02, fontsize=18)

    ### Set the layout and show the plot
    plt.tight_layout()
    plt.show()
    

if __name__ == "__main__":
    sys.exit(main())
```
